```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Market_Basket_Optimisation.csv")
df.info()
```

```python
data=[]
for i in range(len(df)):
  data.append([str(df.values[i,j]) for j in range(20) if str(df.values[i,j]!='0')])
data
```

```python
!pip install apyori
```

```python
from apyori import apriori
rules=apriori(data,min_support=0.003,min_confidence=0.2,min_lift=3,min_length=2)
results=list(rules)
results
```

```python
def fun(results):
  lhs = [tuple(i[2][0][0])[0] for i in results]
  rhs = [tuple(i[2][0][1])[0] for i in results]

  support = [i[1] for i in results]
  confidence = [i[2][0][2] for i in results]
  lift = [i[2][0][3] for i in results]

  return list(zip(lhs,rhs,support,confidence,lift))
data =pd.DataFrame(fun(results),columns=['Item1','Item2','support','confidence','lift'])
data.nlargest(n=10,columns='lift')
```
