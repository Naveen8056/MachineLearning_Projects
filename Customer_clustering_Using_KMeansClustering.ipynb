```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Mall_Customers.csv')
df.drop('CustomerID',axis=1,inplace=True)
df.drop('Genre',axis=1,inplace=True)
df.drop('Age',axis=1,inplace=True)
df
```

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
  km=KMeans(n_clusters=i,init='k-means++',random_state=40)
  km.fit(df)
  wcss.append(km.inertia_)
plt.figure(figsize=(10,5))
plt.plot(range(1,11),wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
sns.boxplot(df)
```

```python
kmean=KMeans(n_clusters=5,init='k-means++',random_state=42)
y=kmean.fit_predict(df)
y
```

```python
kmean.inertia_
```

```python
kmean.labels_
```

```python
df1=df.values
plt.figure(figsize=(15,7))
plt.scatter(df1[y==0,0],df1[y==0,1],color='yellow',label='cluster1',s=50)
plt.scatter(df1[y==1,0],df1[y==1,1],color='red',label='cluster2',s=50)
plt.scatter(df1[y==2,0],df1[y==2,1],color='blue',label='cluster3',s=50)
plt.scatter(df1[y==3,0],df1[y==3,1],color='pink',label='cluster4',s=50)
plt.scatter(df1[y==4,0],df1[y==4,1],color='green',label='cluster5',s=50)
plt.scatter(kmean.cluster_centers_[:,0],kmean.cluster_centers_[:,1],color='black',label='centroids',s=150,marker=',')
plt.grid(False)
plt.title('Clusters of customers')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()
```
