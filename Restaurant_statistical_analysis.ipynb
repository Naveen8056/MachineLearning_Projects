```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_excel('Restaurant.xlsx')
df.head()
b=df[df['INVOICE_NO']==117]
print(b['AMOUNT'].sum())
```

```python
sales=df.groupby('ITEM')['SUB_TOTAL'].sum().sort_values(ascending=False).head(5)
quan=df.groupby('ITEM')['QUANTITY'].sum().sort_values(ascending=False).head(5)
sns.barplot(sales)
sns.barplot(quan)
```

```python
c=df[df['DATE']=='2.10.2023']
least=c.groupby('ITEM')['QUANTITY'].sum().sort_values(ascending=True).head(5)
sns.barplot(least)
```

```python
high=df.groupby('INVOICE_NO')['AMOUNT'].sum().sort_values(ascending=False).head(1)
print(high)
```

```python
data=pd.read_csv('Data.csv')
data.head()
```

```python
a=data.groupby('Date')['Final Total'].mean().sort_values(ascending=False).head(1)
print(a)
```

```python
b=data['Invoice No.'].nunique()
print(b)
```

```python
c=data[(data['Date']>='2023-09-01') & (data['Date']<='2023-09-31')]
d=c.groupby('Item Name')['Qty.'].sum().sort_values(ascending=False).head(1)
print(d)
```

```python
df=data[(data['Date']>='2023-07-01') & (data['Date']<='2023-07-31')]
q=df.groupby('Category')['Qty.'].sum().sort_values(ascending=True).head(3)
sns.barplot(q)
```
